##### ====================================
### 6. Data Cleaning
##### ====================================

```python

# Converting price columns into numeric format
products_df['final_price'] = (
    products_df['final_price']
    .astype(str)
    .str.replace('"', '', regex=False)      
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
    .str.strip()                            
)

products_df['final_price'] = pd.to_numeric(
    products_df['final_price'],
    errors='coerce'
)

products_df['initial_price'] = pd.to_numeric(
    products_df['initial_price'],
    errors='coerce'
)

products_df['discount'] = pd.to_numeric(
    products_df['discount'],
    errors='coerce'
)
```

```python
#Handling Missing Values
numeric_columns = [
    'rating',
    'ratings_count',
    'initial_price',
    'final_price',
    'discount'
]

for column in numeric_columns:
    median_value = products_df[column].median()
    products_df[column] = products_df[column].fillna(median_value)

print(products_df[numeric_columns].isnull().sum())
```

```python
# Handling missing values in text/object columns
text_columns = ['what_customers_said', 'seller_name', 'seller_information', 'videos', 'variations']

for column in text_columns:
    products_df[column] = products_df[column].fillna("Not Available")

print(products_df[text_columns].isnull().sum())
```

```python
products_df.drop(columns=['videos', 'variations'], inplace=True)
```

```python
#Removing Duplicate Records
print("Duplicate Records:", products_df.duplicated().sum())

products_df.drop_duplicates(inplace=True)

print("Dataset Shape After Removing Duplicates:")
print(products_df.shape)

if products_df.duplicated().sum() == 0:
    print("No exact duplicate rows were found in the dataset.")
```

##### ====================================
### 7. Feature Engineering
##### ====================================

```python

products_df['price_difference'] = (
    products_df['initial_price']
    - products_df['final_price']
)

products_df['discount_percentage'] = (
    (products_df['initial_price']
     - products_df['final_price'])
    / products_df['initial_price']
) * 100

products_df['popularity_score'] = (
    products_df['rating']
    * products_df['ratings_count']
)
```

```python
products_df[['price_difference',
             'discount_percentage',
             'popularity_score']].head()
```

**Note:** The provided dataset does not contain a quantity column. Therefore, the required total_amount feature could not be calculated. Instead, additional derived features such as price_difference, discount_percentage, and popularity_score were created to support data analysis and generate meaningful insights.

##### ====================================

### 15. Key Business Insights

##### ====================================

#### Insight 1

Most products have ratings above 4.0 (53.3% of the catalog), indicating a generally high level of customer satisfaction.

**Business Implication:** Products with high ratings can be highlighted in marketing campaigns to attract more customers. Products with ratings below 4.0 may require further review to identify potential quality or customer satisfaction issues.

#### Insight 2

A small number of categories — tops, dresses, shirts, jeans, and sports shoes — account for nearly 43% of the total product catalog.

**Business Implication:** The business is highly focused on a few major categories. Expanding and promoting underrepresented categories may help reduce risk and attract a broader customer base.

#### Insight 3

Products with higher ratings tend to receive more customer reviews, although the relationship is relatively weak (correlation ≈ 0.13). This suggests that customer engagement is influenced by factors beyond product quality alone.

**Business Implication:** Good ratings are important, but product visibility and promotion also play a key role in generating customer reviews and engagement.

#### Insight 4

Higher discount percentages show a mild negative relationship with product ratings (correlation ≈ -0.28), indicating that heavily discounted products tend to have slightly lower ratings.

**Business Implication:** Large discounts should not be considered a replacement for product quality. Categories that rely heavily on discounts may require further investigation to identify possible quality concerns.

#### Insight 5

The popularity score (based on ratings and review volume) helps identify products that are both highly rated and widely reviewed, rather than relying only on ratings.

**Business Implication:** This metric can be used to highlight trending or best-selling products and support recommendation systems more effectively.

#### Insight 6

Categories such as cutlery, shaving essentials, and face wash/cleanser have the highest average ratings, while larger categories such as tops, dresses, and shirts remain close to the overall average.

**Business Implication:** Smaller categories may contain highly appreciated products and could offer opportunities for additional promotion. This may also help category managers identify areas for further improvement in larger product segments.
